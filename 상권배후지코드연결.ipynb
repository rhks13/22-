```python
# 드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import os
from glob import glob
import pickle
with open('/content/drive/MyDrive/통계 빅데이터 공모전/data/상권 특성/상권배후지data.pickle', 'rb') as f:
    df = pickle.load(f) 
상권 = pd.read_excel("/content/drive/MyDrive/통계 빅데이터 공모전/data/상권 특성/상권리스트.xlsx", )
trdar = 상권.copy()
geom=trdar['geometry'].str.split(' ')
l1=[]

for j in range(len(geom[0])):
  l1.append(geom[0][j].split(','))
l2 = pd.DataFrame(l1).iloc[:,0]
l2 = l2.astype(float)
y_max = l2[1::2].max()
y_min = l2[1::2].min()
x_max = l2[::2].max()
x_min = l2[::2].min()
#초기 데이터프레임 생성
ran = [x_min, x_max, y_min, y_max]
df11 = pd.DataFrame(ran).T
for i in range(1,len(geom)): 
  l1=[]
  
  for j in range(len(geom[i])):
    l1.append(geom[i][j].split(','))
  l2 = pd.DataFrame(l1).iloc[:,0]
  try:
    l2 = l2.astype(float) 
  except:
     pass
  y_max = l2[1::2].max() 
  y_min = l2[1::2].min() 
  x_max = l2[::2].max() 
  x_min = l2[::2].min() 
  ran = [x_min, x_max, y_min, y_max]
  df22 = pd.DataFrame(ran).T
  df11 = df11.append(df22)
df11.index=trdar['TRDAR_NO']
df11['num']=range(len(df11))
df11[0]=df11[0].astype(float)-200
df11[1]=df11[1].astype(float)+200
df11[2]=df11[2].astype(float)-200
df11[3]=df11[3].astype(float)+200
with open('/content/drive/MyDrive/통계 빅데이터 공모전/data/업체 데이터/데이터최종원본.pickle', 'rb') as f:
    main_df = pickle.load(f) 
main_data = main_df.copy() 
main_data['상권코드'] = 0
main_data['번호']=range(len(main_data))
def trdar(i):
  try:
    x=main_data['좌표정보(X)'].astype(float)
    y=main_data['좌표정보(Y)'].astype(float)
    return np.where((x>=df11.iloc[i,0])&(x<=df11.iloc[i,1])&(y>=df11.iloc[i,2])&(y<=df11.iloc[i,3]))
  except:
    pass
df11['상권'] = df11['num'].map(lambda x : trdar(x))
#빈 리스트에 모두 추가하는 방식
f_list=[]
k_=[]
for i in range(len(df11)):
  try:
    for j in range(len(df11['상권'].iloc[i][0])):
       f_list.append(df11['상권'].iloc[i][0][j])
    k_+=[f_list]
  except:
    pass  
ind_ = list(map(lambda x: list(list(x)[0]) ,df11.상권))
cod_ = list(df11.index)
def map_(x):
  main_data['상권코드'].loc[ind_[x]]=cod_[x]

list(map(map_, range(len(ind_))))
sub_data = main_data[main_data==0]
with open('/content/drive/MyDrive/통계 빅데이터 공모전/data/상권 특성/상권배후지코드연결.pickle', 'wb') as f:
    pickle.dump(sub_data, f, pickle.HIGHEST_PROTOCOL)
```
